```python
outlier_cols = [
    'TopographyDrainage',
    'Urbanization',
    'DrainageSystems'
]

for col in outlier_cols:
    Q1 = df1[col].quantile(0.25)
    Q3 = df1[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_limit = Q1 - (1.5 * IQR)
    upper_limit = Q3 + (1.5 * IQR)

    df1[col] = df1[col].clip(lower=lower_limit, upper=upper_limit)
```

```python
x=df1.iloc[:,:-1]
x
```

```python
y=df1.iloc[:,-1]
y
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.2, random_state=42
)
```

```python
df1.reset_index(drop=True, inplace=True)
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score


rf_model = RandomForestRegressor(
    n_estimators=100,
    random_state=42
)

rf_model.fit(x_train, y_train)

rf_pred = rf_model.predict(x_test)

```

```python
rf_pred
```

```python
MAE=mean_absolute_error(y_test,rf_pred)
MAE
```

```python
MSE=mean_squared_error(y_test,rf_pred)
MSE
```

```python
from sklearn.metrics import mean_absolute_percentage_error
MAPE=mean_absolute_percentage_error(y_test,rf_pred)
MAPE
```

```python
R2S=r2_score(y_test,rf_pred)
R2S
```

```python
print("Training Score :", rf_model.score(x_train, y_train))
print("Testing Score  :", rf_model.score(x_test, y_test))
```

```python
rf_result = pd.DataFrame({
    'Actual': y_test,
    'Predicted': rf_pred,
    'Difference': y_test - rf_pred
})

rf_result.head()
```

```python
from xgboost import XGBRegressor

xgb_model = XGBRegressor(
    n_estimators=100,
    random_state=42
)

xgb_model.fit(x_train, y_train)

xgb_pred = xgb_model.predict(x_test)
```

```python
xgb_pred
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, mean_absolute_percentage_error, r2_score,root_mean_squared_error


XMAE = mean_absolute_error(y_test, xgb_pred)
XMSE = mean_squared_error(y_test, xgb_pred)
XRMSE = root_mean_squared_error(y_test, xgb_pred)
XMAPE = mean_absolute_percentage_error(y_test, xgb_pred)
XR2 = r2_score(y_test, xgb_pred)

print("XMAE :", XMAE)
print("XMSE :", XMSE)
print("XRMSE:", XRMSE)
print("XMAPE:", XMAPE)
print("XR2 Score:", XR2)
```

```python
print("Training Score :", xgb_model.score(x_train, y_train))
print("Testing Score  :", xgb_model.score(x_test, y_test))
```

```python
xgb_result = pd.DataFrame({
    "Actual": y_test,
    "Predicted": xgb_pred,
    "Difference": y_test - xgb_pred
})

xgb_result.head()
```

```python
models = ['Random Forest', 'XGBoost']
scores = [R2S, XR2]

plt.bar(models, scores,color=['g','orange'])

for i in range(len(scores)):
    plt.text(i, scores[i] + 0.01, round(scores[i], 2), ha='center')

plt.xlabel("Models")
plt.ylabel("R2 Score")
plt.title("Model Performance Comparison")
plt.ylim(0, 1)
plt.show()

```
